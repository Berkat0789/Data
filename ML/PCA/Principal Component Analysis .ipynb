### WEw ill use a PCA aproach on breast cancer data
-What is a traisnformaton of data in the atempt whihc feature explains the most Varience of the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer["data"], columns=cancer["feature_names"])
```

```python
df.corr()
```

### WE need to find th most important components in the data set
- WE will not be doing classificaiton or predicitons
- We have 30 featues and dimansions 
- WE wcan use PCA to get an idea of what is important -> Difficult to visualize
- WE will use PCA to fince the first 2 PCA

```python
from sklearn.preprocessing import StandardScaler
```

### scaling the data
- So that each feature has a single unit varience

```python
scalar = StandardScaler()
```

```python
scaled_data = scalar.fit_transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
principal_C = PCA(n_components=2)
```

```python
X_pca = principal_C.fit_transform(scaled_data)
```

```python
X_pca
```

```python
scaled_data.shape
```

```python
X_pca.shape
```

```python
X_pca_df = pd.DataFrame(X_pca)
```

```python
X_pca_df["target"] = cancer["target"]
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(X_pca_df[0], X_pca_df[1], data=X_pca_df, hue="target")
```

### PCas can be used a a sort of compresssion analysis-. So the 2 pCA that we Chose actually have a high prediciton on the Benign and Melignant target. So we can use these two components to do our analysis- Instead of 30 columns of data

### First component and Second Components 
- unfortunately if comes with a cost of being able to understand the featues of the data
- iT is a combination of the features 
- components is stored an an sttribut of the PCA object
- Each row relates to the pCA and each colum relates back to the acrual features

```python
principal_C.components_
```

```python
comp_df = pd.DataFrame(principal_C.components_, columns=cancer["feature_names"])
```

```python
comp_df
```
