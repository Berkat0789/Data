```python
df["gender"] = pd.get_dummies(df["gender"], drop_first=True)
```

```python
df["ever_married"] = pd.get_dummies(df["ever_married"], drop_first=True)
```

```python
work_type = pd.get_dummies(df["work_type"], drop_first=True)
```

```python
df = pd.concat([df,work_type], axis=1)
```

```python
df.drop("work_type", axis=1, inplace=True)
```

```python
df["Residence_type"]=pd.get_dummies(df["Residence_type"], drop_first=True)
```

```python
smok = pd.get_dummies(df["smoking_status"], drop_first=True)
```

```python
df = pd.concat([df, smok], axis=1)
```

```python
df.drop('smoking_status', axis = 1, inplace=True)
```

## Data is prepared
- Now that we have all numerical data

```python
X = df.drop("stroke", axis=1).values
y = df["stroke"].values
```
